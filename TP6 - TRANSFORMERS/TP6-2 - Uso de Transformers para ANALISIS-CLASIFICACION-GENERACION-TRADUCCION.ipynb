En el siguiente ejemplo, preguntamos al modelo qu茅 desea el cliente, utilizando el pipeline `question-answering`.

```python
# Modelo de respuesta a preguntas en espa帽ol
# PlanTL-GOB-ES/roberta-large-bne-sqac es de PlanTL-GOB-ES, entrenado en SQAC.
reader = pipeline("question-answering", model="PlanTL-GOB-ES/roberta-large-bne-sqac")
```

## 5. Resumen autom谩tico de texto

El **resumen autom谩tico** permite condensar la informaci贸n principal de un texto extenso en una versi贸n m谩s corta, manteniendo el significado esencial. Los Transformers han revolucionado esta tarea gracias a su capacidad de comprensi贸n contextual.

A continuaci贸n, utilizamos el pipeline `summarization` para obtener un resumen del texto de ejemplo.

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
```

```python
# Cargar modelo y tokenizer
model_name = "csebuetnlp/mT5_multilingual_XLSum"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)
```

```python
# Crear el pipeline
resumidor = pipeline("summarization", model=model, tokenizer=tokenizer)
```

```python
# Texto para resumir
texto = """En el Mundial de Clubes, el Manchester City de Inglaterra se enfrent贸 al Fluminense de Brasil en una final emocionante.
Juli谩n lvarez fue figura clave para los ingleses, marcando un gol decisivo.
Por otro lado, Marcelo, exjugador del Real Madrid, lider贸 la defensa del equipo brasile帽o.
Desde Arabia Saudita, sede del torneo, tambi茅n destacaron clubes como el Urawa Red Diamonds de Jap贸n y el Al Ahly de Egipto.
Cada partido mostr贸 el talento internacional reunido en esta prestigiosa competencia."""
```

```python
# Generar resumen
resumen = resumidor(texto, max_length=80, min_length=20, do_sample=False)
print(resumen[0]['summary_text'])
```

## 7. Generaci贸n autom谩tica de texto (respuesta de servicio al cliente)

La **generaci贸n autom谩tica de texto** permite crear o continuar textos de manera coherente a partir de un prompt inicial. Esta capacidad es 煤til, por ejemplo, para redactar respuestas autom谩ticas en atenci贸n al cliente.

En el siguiente ejemplo, generamos una posible respuesta de servicio al cliente utilizando el pipeline `text-generation`.

```python
# Modelo de generaci贸n de texto en espa帽ol
# DeepESP/gpt2-spanish es un GPT-2 entrenado desde cero con texto y un tokenizer en espa帽ol.
generator = pipeline("text-generation", model="DeepESP/gpt2-spanish")
```

```python
response = "Estimado Bumblebee, lamento mucho saber que su pedido se mezcl贸."
prompt = texto + "\n\nRespuesta del servicio al cliente:\n" + response
outputs = generator(prompt, max_length=100, do_sample=True, temperature=1.0) # A帽adir par谩metros para mejor generaci贸n
print(outputs[0]['generated_text'])
```

```python
from transformers import pipeline

# Modelo en espa帽ol
generador = pipeline("text-generation", model="datificate/gpt2-small-spanish")

# Texto del cliente
texto = """En el Mundial de Clubes, el Manchester City de Inglaterra se enfrent贸 al Fluminense de Brasil en una final emocionante.
Juli谩n lvarez fue figura clave para los ingleses, marcando un gol decisivo.
Por otro lado, Marcelo, exjugador del Real Madrid, lider贸 la defensa del equipo brasile帽o.
Desde Arabia Saudita, sede del torneo, tambi茅n destacaron clubes como el Urawa Red Diamonds de Jap贸n y el Al Ahly de Egipto.
Cada partido mostr贸 el talento internacional reunido en esta prestigiosa competencia."""

# Primer rengl贸n de la respuesta
respuesta_inicial = "Estimado Bumblebee, lamentamos mucho lo ocurrido con su pedido. "

# Prompt claro, sin repetir encabezado
prompt = texto + "\n\nRespuesta del servicio al cliente:\n" + respuesta_inicial

# Generaci贸n
outputs = generador(
    prompt,
    max_new_tokens=150,
    do_sample=True,
    temperature=1.0,
    top_k=50,
    top_p=0.9,
    repetition_penalty=1.3,
    eos_token_id=50256  # token de final en GPT-2
)

# Mostrar resultado
print(outputs[0]['generated_text'])
```
